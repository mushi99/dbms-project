```python
import requests

url = 'https://gbfs.citibikenyc.com/gbfs/en/station_status.json'
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    station_status_data = data['data']['stations']

else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
def query_without_index(collection):
    start_time = time.time()

    query_ref = collection.where('num_bikes_available', '>', 2).order_by("num_docks_available", direction=firestore.Query.ASCENDING)
    docs = query_ref.stream()
    
    end_time = time.time()
    execution_time = end_time - start_time
 
    print(f"Execution time: {execution_time} seconds")
    return execution_time
```

```python
index_ref = db.collection('citi-bike').create_index([firestore.FieldPath('num_bikes_available', 'num_docks_available')])
print("Composite index created successfully.")
```
